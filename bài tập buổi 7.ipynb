```python
#W7A1#
arr=list(map(int,input().split()))
target=int(input())
def check(arr,l,r,target):
    while l<=r:
        mid=(l+r)//2
        if arr[mid]==target: return mid
        elif arr[mid]<target:
            l=mid+1
        else:
            r=mid-1
    return -1
check(arr,0,len(arr)-1,target)
```

```python
#W7A2#
arr=list(map(int,input().split()))
arr.sort()
target=int(input())
def check(arr,target):
    def checkleft(arr,target)->int:
        l,r=0,len(arr)-1
        while l<=r:
            mid=(l+r)//2
            if arr[mid]>=target: r=mid-1
            else: l=mid+1
            if arr[l]==target: return l
        return 0
    def checkright(arr,target)->int:
        l,r=0,len(arr)-1
        while l<=r:
            mid=(l+r)//2
            if arr[mid]<=target: l=mid+1
            else: r=mid-1
            if arr[r]==target: return r
        return 0
    return checkright(arr,target)-checkleft(arr,target) if checkright(arr,target)!=0 and checkleft(arr,target)!=0 else 0
print(check(arr,target))
```

```python
#W7A3#

arr=list(map(int,input().split()))

def sortBB(arr):
    for i in range(len(arr)):
        ans=True
        for j in range(0,len(arr)-1):
            tempt=arr[j]
            if arr[j]>arr[j+1]:
                ans=False
                arr[j]=arr[j+1]
                arr[j+1]=tempt
        if ans: return arr
    return arr

def sortSLT(arr:list)->list:
    for i in range(len(arr)-1):
        tempt=arr[i]
        minn=arr[i+1]
        indexx=i+1
        for j in range(i+1,len(arr)):
            if arr[j]<minn:
                minn=arr[j]
                indexx=j
        if arr[i]>minn:
            arr[i]=minn
            arr[indexx]=tempt
    return arr
print(sortSLT(arr))
print(sortBB(arr))
                   
```
