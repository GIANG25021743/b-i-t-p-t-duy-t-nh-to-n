```python
#W7A1#
arr=list(map(int,input().split()))
target=int(input())
def check(arr,l,r,target):
    while l<=r:
        mid=(l+r)//2
        if arr[mid]==target: return mid
        elif arr[mid]<target:
            l=mid+1
        else:
            r=mid-1
    return -1
check(arr,0,len(arr)-1,target)
```
