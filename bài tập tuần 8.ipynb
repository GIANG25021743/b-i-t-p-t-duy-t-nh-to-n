```python
#w8A5
class ShoppingCart:
    def __init__(self, name , price):
        self.n=name
        self.p=price
    gio_hang={}
    def add(self):
        ShoppingCart.gio_hang[self.n]={"price":self.p}
    def del_item():
        item=input("nhập tên sản phẩm bạn muốn gỡ:")
        if item in ShoppingCart.gio_hang: 
            del  ShoppingCart.gio_hang[item] 
        else: print("Không tìm thấy sản phẩm!")
    def check_empty():
        if len(ShoppingCart.gio_hang)==0: print("chưa có sản phẩm nào")
        if len(ShoppingCart.gio_hang)!=0: print(f"{len(ShoppingCart.gio_hang)} sản phẩm")    
    def summ():
        sum=0
        for item in ShoppingCart.gio_hang:
            sum+=ShoppingCart.gio_hang[item]["price"]
        print(sum)
    def all():
        for item in ShoppingCart.gio_hang:
            print(f"{item}   :   {ShoppingCart.gio_hang[item]["price"]}")
    def del_all():
        ShoppingCart.gio_hang=[]
print("=======MENU=======")
print("""các phương thức cho khách hàng:
            2-> add new item
            3-> remove item
            4-> check empty
            5-> calculate the sum of all item
            6-> display all item
            7-> delete shopping cart""")
tempt=input("chọn yêu cầu của bạn (điền số):")
customer=ShoppingCart("none",0)
while True:
    if tempt=="2":
        n=input("nhập tên sản phẩm")
        p=float(input(("nhập giá sản phẩm")))
        customer=ShoppingCart(n,p)
        customer.add()
    elif tempt=="3":
        ShoppingCart.del_item()
    elif tempt=="4":
        ShoppingCart.check_empty()
    elif tempt=="5":
        ShoppingCart.summ()
    elif tempt=="6":
        ShoppingCart.all()
    elif tempt=="7":
        ShoppingCart.del_all()
    elif tempt=="stop":
        break
    tempt=input("chọn yêu cầu của bạn (điền số):")
```

```python
#W8A6#
class Rectangle:
    def __init__(self, w ,h):
        self.w=w
        self.h=h
    def area(self):
        print(f"the area equals {(self.w*self.h):.2f}")
    def perimeter(self):
        print( f"perimeter equals {((self.w+self.h)*2):.2f}")
    def scale(self,k):
        self.w*=k
        self.h*=k
a,b,c=map(float,input().split())
tempt=Rectangle(a,b)
tempt.scale(c)
tempt.area()
tempt.perimeter()
```

```python
#w8a8#
class BankAccount:
    def  __init__(self,owner,balance):
        self.rest=float(balance)
        self.name=owner
    def Deposite(self,amount):
        self.rest+=amount
    def Withdraw(self,amount):
        if self.rest>=amount:
            self.rest-=amount
    def __str__(self):
        return f"số tiền còn lại trong tài khoản:{self.rest:.2f}"
a,b=map(str,input().split())
customer=BankAccount(a,b)
n=int(input("số lần thực hiện:"))
for i in range(n):
    offer=input()
    a,b=map(str,offer.split())
    if "DEPOSITE" in offer:
        customer.Deposite(float(b))
    elif "WITHDRAW" in offer:
        customer.Withdraw(float(b))
print(customer.__str__())
```

```python
#w8A10#
class Book:
    def __init__(self,title,author,year):
        self.t=title
        self.n=author
        self.y=year
class Library(Book):
    book_container=[]
    def __init__(self,title,author,year):
        super().__init__(title,author,year)
    def add(self):
        Library.book_container+=[{"title":self.t,"author":self.n,"year":self.y}]
    @classmethod
    def count_by_author(cls,a):
        cls.check_author=a
        count=0
        for book in Library.book_container:
            if book["author"]==cls.check_author:
                count+=1
        return count
    @classmethod
    def countyear(cls,b):
        cls.check_year=b
        count=0
        for book in Library.book_container:
            if book["year"]==cls.check_year:
                count+=1
        return count
n=int(input("số lần thực hiện"))
for i in range(n):
    tempt=input("nhập yêu cầu của bạn:")
    if "ADD" in tempt:
        a,b=map(str,tempt.split())
        title,author,year=map(str,b.split(";"))
        item=Library(title,author,int(year))
        item.add()
    elif "COUNTAUTHOR" in tempt:
        a,b=map(str,tempt.split())
        print(Library.count_by_author(b))
    elif "COUNTYEAR" in tempt:
        a,b=map(str,tempt.split())
        print(Library.countyear(int(b)))
        
```

```python
#w8a12#
class ComplexNumber:
    def __init__(self,realpart,imaginarypart):
        self.real=realpart
        self.ima=imaginarypart
    @classmethod
    def multiply(cls,a,b):
        cls.real=a.real*b.real - a.ima*b.ima
        cls.ima=a.real*b.ima + a.ima*b.real
        print(f"newcomplexnumber: {cls.real}  {cls.ima}")
tempt1=ComplexNumber(1,2)
tempt2=ComplexNumber(3,4)
ComplexNumber.multiply(tempt1,tempt2)
#nếu không muốn dùng cls thì bỏ @classmethod đi#
```

```python
#w8a13#
class Triangle:
    def __init__(self, a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def check(self):
        return self.a+self.b>self.c and self.a+self.c>self.b and self.c+self.b>self.a
    def getArea(self):
        if self.check():
            self.p=(self.a+self.b+self.c)/2
            area=(self.p*(self.p-self.a)*(self.p-self.b)*(self.p-self.c))**0.5
            return area
        else: 
            return -1 
    def __str__(self):
        if self.getArea()!=-1:
            print(f"diện tích của tam giác là {self.getArea():.2f}")
        else: 
            print("invalid")
a,b,c=map(float,input("nhập độ dài 3 cạnh của tam giác:").split())
tempt=Triangle(a,b,c)
tempt.__str__()
```
