```python
n=int(input("nhập một số dương:"))
def check(n):
    if n==0 or n==1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
def uoc(n):
    dem=n
    while dem>0:
        if n%dem==0 and check(dem):
            break  
        dem-=1
    return dem
print(uoc(n))
```

```python
def so_dao_nguoc(n):
    b=0
    while n>0:
        b= b*10 + n%10
        n= n//10
    return b
def check(n):
    count=0
    while n!=so_dao_nguoc(n):
        n += so_dao_nguoc(n)
        count += 1
        
    return {count,n}
n=int(input())
print(check(n))
```
