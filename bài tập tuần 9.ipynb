```python
#w9a5#
a=list(map(int,input().split()))
b=list(map(int,input().split()))
def check(a,b)->bool:
    i,j=0,0
    while i <len(b) and j <len(a):
        if a[j]==b[i]:
            if j==len(a)-1:
                return True
            else:
                j+=1
        elif a[j]!=b[i]:
            j=0
        i+=1
    return False
print(check(a,b))
```

```python
#w9a7#
k=int(input())
space=list(map(int,input().split()))
left=True
right=True
for i in range(len(space)):
    if i==0 or (i!=0  and space[i-1]!=1):
        left=True
    elif i!=0 and space[i-1]==1:
        left=False
    if i==len(space)-1 or (i!=len(space)-1 and space[i+1]!=1):
        right=True
    elif i!=len(space)-1 and space[i+1]==1:
        right=False
    if space[i]==0 and left and right:
        space[i]=1
        k-=1
if k==0:
    print("YES")
else:
    print("NO")
 
```

```python
#w9a8#
def is_path_crossing(moves:str)->bool:
    x,y=0,0
    check=[[0,0]]
    for move in moves:
        if move=="R":
            x+=1
        elif move=="L":
            x-=1
        elif move=="U":
            y+=1       
        elif move=="D":
            y-=1
        if [x,y] in check:
            return True
        check.append([x,y])
    return False
moves=input()
print(is_path_crossing(moves))

        
```
