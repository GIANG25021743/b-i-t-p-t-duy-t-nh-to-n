```python
#W6a9#
dict1={'b':3,'c':4,"a":4}
dict2={'a':3,'d':5,'c':1}
def check(a: dict,b:dict)-> dict:
    tempt=a
    for i in b:
        if i not in tempt:
            tempt[i]=b[i]
        else: 
            tempt[i]+=b[i]
    tempt1={}
    giang=list(tempt.keys())
    giang.sort()
    for i in giang:
        tempt1[i]=tempt[i]
    return tempt1
print(check(dict1,dict2))
```

```python
#W6a10#
a=list(map(int,input().split()))
k=int(input())
def check(a,k) -> list:
    tempt=[]
    for i in range(len(a)):
        for j in range(i+1,len(a)):
            if a[i]+ a[j]==k:
                if (a[i],a[j]) not in tempt:
                    tempt.append((a[i],a[j]))
    return tempt
print(check(a,k))
```

```python
#W6a12#
a=list(map(int,input().split()))
def check(a) -> int:
    dicttempt={}
    maxx=0
    for i in a:
        if i not in dicttempt : dicttempt[i]=1
        else : 
            dicttempt[i]+=1
        if dicttempt[i]>maxx : 
            maxx=dicttempt[i]
            minn=i
        if dicttempt[i]==maxx and i<minn:
            minn=i
    return minn
print(check(a))
```

```python
#W6a13#
a=list(map(str,input().split()))
def check(a) -> dict:
    dict_tempt={}
    i=0
    while i<len(a)-1:
        dict_tempt[a[i]]=a[i+1]
        i+=2
    newdict={}
    for i in dict_tempt:
        newdict[dict_tempt[i]]=i
    return newdict
print(check(a))
```

```python
#W6a14#
a=list(map(int,input().split()))
b=list(map(int,input().split()))
def check(a,b)-> list:
    tempt=[]
    dict1={}
    dict2={}
    for i in a:
        if i not in dict1:
            dict1[i]=1
        else:
            dict1[i]+=1
    for j in b:
        if j not in dict2:
            dict2[j]=1
        else:
            dict2[j]+=1
    for i in dict1:
        if i in dict2:
            tempt.append(i)
    return tempt
print(check(a,b))
       
```

```python
#W6a15#
a=list(map(str,input().split()))
k=int(input())
def check(a,k) -> dict:
    i=0
    dict_tempt={}
    while i<len(a)-1:
        if int(a[i+1])>k:
            dict_tempt[a[i]]=int(a[i+1])
        i+=2
    return dict_tempt
print(check(a,k))
```

```python
#W6a19#
a=list(map(int,input().split()))
def check(a) ->int:
    maxx=a[0]
    tempt=0
    for loca,num in enumerate(a):
        if num>maxx:
            maxx=num
            tempt=loca
    return tempt+1
print(check(a))
```
